```python
name="Almıla Sultan"
surname="TAŞ"
iname=input("Please enter your name: ").strip()
isurname=input("Please enter your surname: ").strip()
chance=2
check=False
while chance>0:
    if name.lower()==iname.lower() and surname.lower()==isurname.lower():
        print(f"Welcome {name} {surname}.")
        chance=0
        check=True
    else:
        print("Try again...")
        iname=input("Please enter your name: ")
        isurname=input("Please enter your surname: ")
        chance=chance-1
        if chance==0:
            print("Please try again later.")
if check==True:
    lessons=[]
    i=0
    while i<5:
        lesson=input("Please enter your lesson: ")
        if len(lesson)>0:
            lessons.append(lesson)
        i=i+1
    if len(lessons)<3:
        print("You failed in class.")
    else:
        print("Choose one of this lessons.")
        for i in range(len(lessons)):
            
            print(i+1, " ", lessons[i])
            
        choose=int(input("Lesson's number: "))
        if choose>len(lessons):
            choose=int(input("Please try again enter lesson's number: "))
            
        print(f"Choosed lesson is {lessons[choose-1]}")
        midterm=int(input("Please enter your miderm grade: "))
        if midterm>100:
            midterm=int(input("Please enter again your miderm grade: "))
        final=int(input("Please enter your final grade: "))
        if final>100:
            final=int(input("Please enter again your final grade: "))
        project=int(input("Please enter your project grade: "))
        if project>100:
            project=int(input("Please enter again your project grade: "))
        dicgrade={"midterm":midterm, "final":final, "project":project}
        print (dicgrade)
        grade=(dicgrade["midterm"]*0.3)+(dicgrade["final"]*0.5)+(dicgrade["project"]*0.2)
        print(f"Your grade is {grade}")
        if grade >=90:
            print("You passed with AA")
        elif grade <90 and grade>=70:
            print("You passed with BB")
        elif grade <70 and grade>=50:
            print("You passed with CC")
        elif grade <50 and grade>=30:
            print("You passed with DD")
        elif grade <30 :
            print("You failure with FF")
```
